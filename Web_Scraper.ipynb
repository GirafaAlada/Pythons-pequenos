## Web Scraper basico

```python
# Nesse exemplo vou escrever um código simples comentado só pra manter guardado o básico ara poder fazer funcionar o scraper.

# Aqui importamos as bibliotecas que serão usadas
from bs4 import BeautifulSoup
import requests


# Esse é o endereço que vou usar.
# ALGUMA COISA MUDOU NO SITE E O CÓGIGO ESTÁ QUEBRADO
url = "https://www.pichau.com.br/controle-gamer-tgt-ac130-pc-ps3-tgt-ac130"

# Aqui eu pego o site e crio uma variavel que contem o texto do codigo html
result = requests.get(url)
doc = BeautifulSoup(result.text, "html.parser")

# Esse prettify é pra fazer a endentação pra facilitar a leitura, mas não precisa
print(doc.prettify())

# Agora um codigo que procura um termo no texto que me ajude a localizar a infromação que eu preciso
# Nesse caso eu escolhi o termo R$ já que eu queria esse valor
# o find_all vai listar em uma variável todas as ocorrencias
# Por sorte minha só tem uma ocorrencia aqui
preços = doc.find_all(text="R$")
print(preços)

# Essa nova variável pega a tag onde tem o termo que eu localizei anteriormente
tagPai = preços[0].parent
print(tagPai)

# Eu escolhi transformar essa variavel em uma string pra facilitar trabalhar com ela
print("O tipo de variável do Beautiful Soup é esse:", type(tagPai))
valor = str(tagPai)
print("Aqui o tipo que eu quero:", type(valor))


# Agora esse loop encontra o indice do termo que eu vou usar pra tirar o preço que eu preciso de dentro da string
# Eu sei que tem uma exclamação um pouco antes do valor, então vou usar isso.
# Poderia ser o cifrão, a barra do fechamento da tag do html ou então a vírgula. Vai o que achar mais fácil
index = []
for n in valor:
    if (n == "!"):
        index = valor.index(n)
print("Aqui o indice da exclamação:", index)

# E por último eu pego o valor que preciso baseado no indice encontrado anteriormente
preco_do_produto = "R$" + valor[index + 7] + valor[index + 8] + valor[index + 9] + valor[index + 10] + valor[index + 11]
print("O preço do treco é: ", preco_do_produto)

# Pronto. Podemos fazer isso com um monte de páginas.
# Talvez esse código pare de funcionar caso acabe o produto na loja. Se for o caso é só trocar o url por outro
# produto da pichau.
# Acredito que vá funcionar tudo normalmente.
```

# Gerando um grafico

```python
# Vamos importar uma bibliioteca que faz graficos
import matplotlib.pyplot as plt
import random

# Criando o eixo x e y:
# Aqui estão as notas que vão ser o eixo y
i = 1
notas_matematica = []
while i < 11:
  notas_matematica.append(random.randrange(1, 10))
  i = i + 1
y = notas_matematica
print("Essas são as notas:", y)

# Agora o eixo x:
x = []
e = 1
while e <= len(y):
  x.append(e)
  e = e + 1

# Agora sim desenhar o grafico com p matplot:
plt.plot(x, y, marker="o")
plt.title("Grafico das notas")
plt.xlabel("Atividades")
plt.ylabel("Notas")
plt.show()
```

## Tratando os dados

```python
from bs4 import BeautifulSoup
import pandas
from urllib.request import Request, urlopen

url = "https://alura-site-scraping.herokuapp.com/hello-world.php"
resposta = urlopen(url)
html = resposta.read()



# O código está na classe Bytes
print(type(html))

# Vamos usar um decodificador pra conseguir ler os caracteres especiais
# Agora ele virou uma str
html = html.decode("utf-8")
print(type(html))

# Podemos transformar o conteudo da str em uma lista
html = html.split()
print(html)

# Eu acho que daria para usar o conteúdo assim já
# Mas podemos transformar tudo em texto novamente com join()
html = " ".join(html)
print(html)
# Agora temos um texto mais limpo sem aqueles /n etc.

# As vezes o Beautiful Soup se confunde quando ve espaços entre as tags
# Então vamos substituir
html = html.replace("> <", "><")
print(html)

# Pronto! Agora ficou tudo bonito
```

### Eu posso agora fazer uma função que já faça tudo isso pra mim

```python
def trata_codigo(codigo):
  return " ".join(codigo.decode("utf-8").split()).replace("> <", "><")

url_copel = "https://statusinvest.com.br/acoes/cple6"

# Deu erro 404 então precisa das informações dos headers
headers_copel = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36'}

req = Request(url_copel, headers = headers_copel)
resposta_copel = urlopen(req)
html_copel = resposta_copel.read()

html_copel = trata_codigo(html_copel)
print(html_copel)
```

### Transformando em um objeto do Beautiful Soup

```python
soup = BeautifulSoup(html, "html.parser")
soup_copel = BeautifulSoup(html_copel, "html.parser")

print(soup.html.head.title)
# Para facilitar a leitura do código, podemos usar o prettify()
print(soup.prettify())
print(soup_copel.prettify())
```
